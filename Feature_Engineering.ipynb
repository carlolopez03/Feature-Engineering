```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler,OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.impute import SimpleImputer

from sklearn import set_config
set_config(transform_output='pandas')
pd.set_option('display.max_columns',100)
```

```python
bikeshare = pd.read_csv('Data/bikeshare_train - bikeshare_train.csv')
bikeshare.info()
bikeshare.head()
```

```python
bikeshare = bikeshare.drop(['casual', 'registered'], axis=1)
bikeshare.head(3)
```

```python
bikeshare['datetime'] = pd.to_datetime(bikeshare['datetime'])
bikeshare.info()
```

```python
bikeshare['month'] = bikeshare['datetime'].dt.month_name()
bikeshare['day of the week'] = bikeshare['datetime'].dt.day_name()
bikeshare['hour of the day'] = bikeshare['datetime'].dt.hour
bikeshare.head()
```

```python
bikeshare = bikeshare.drop(['datetime', 'season'], axis=1)
bikeshare.head(3)
```

```python
bikeshare[['temp','atemp']] = bikeshare[['temp','atemp']].apply(lambda x: (x * 1.8) + 32)
bikeshare.head()
```

```python
bikeshare['temp variance'] = bikeshare['temp'] - bikeshare['atemp']
bikeshare = bikeshare.drop(['atemp'], axis=1)
bikeshare.head()
```
